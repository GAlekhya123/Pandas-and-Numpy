```python
#joining
a = pd.DataFrame({"id":[1,2],"name":["ale","leka"],"des":["ass","log"],"gender":["m","f"]})
b = pd.DataFrame({"id":[1,2],"name":["alfe","lefka"],"des":["afss","lfog"],"gender":["m","m"]})
```

```python
pd.merge(a,b,on="id")
```

```python
pd.merge(a,b,on=["id","gender"])
```

```python
#concat
a = pd.DataFrame({"id":[1,2],"name":["ale","leka"],"des":["ass","log"],"gender":["m","f"]})
b = pd.DataFrame({"id":[1,2],"age":["alfe","lefka"],"de":["afss","lfog"],"ge":["m","m"]})
```

```python
pd.concat([a,b],axis=1)
```
